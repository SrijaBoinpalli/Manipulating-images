```python
croppedIm = catIm.crop((135, 145, 265, 260))
croppedIm.save('cropped.png')
croppedIm
```

```python
croppedIm.size
```

```python
catIm = Image.open('cat.png')
catCopyIm = catIm.copy()
catCopyIm.paste(croppedIm, (0, 0))
catCopyIm.paste(croppedIm, (200, 300))
catCopyIm.save('pasted.png')
catCopyIm
```

```python
catImWidth, catImHeight = catIm.size
croppedImWidth, croppedImHeight = croppedIm.size
catCopyTwo = catIm.copy()
```

```python
for left in range(0, catImWidth, croppedImWidth):
    for top in range(0, catImHeight, croppedImHeight):
        print(left, top)
        catCopyTwo.paste(croppedIm, (left, top))
```
